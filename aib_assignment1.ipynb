## Churn of Members

### Churn Rate Calculations

Churn rate is the number of customers who terminate their relationship with a business in a given time period. It has been calculated by finding the number of customers who left at the end of a time period and dividing it by the number of customers who were present at the beginning of the time period.[1]

##### Example


For the month of December in 2015, Dec-2015,

###### Total Members at the beginning of the Time Period: 

Total Member(Nov-2015) = Total Member(Oct-2015) + Join Member(Nov-2015) – End Member(Nov-2015)

                       = 37591 + 8929 - 8399 
                       = 38121

###### Total Members who leave at the end of the time period: 

Members who end membership in December 2015 = 18165

###### Churn Rate Percentage

Churn rate = End Member(Dec-2015) / Total Member (Nov-2015)

= 18165/38121 = 47.65%

### Key KPIs 

The following are key KPIs that OC&Gym should optimize in order to achieve their objective. 

- Monthly churn rate: calculated as the total number of customers who ended the relationship with the business in a specified month divided by the total number of customers at the beginning of the month. 

- Net promotor score (NPS): a metric used to gauge customer satisfaction by asking them to rate on a scale of 10, whether or not they would recommend the business to their friends. Score that ranges between 9 to 10 are classified as "promoters", 7 to 8 as "passives", and 0 to 6 as "Detractors." 

- Peak time to off-peak time demand ratio: calculated as number of visits during peak time divided by the number of visits during off-peak time. 

- Number of visits per customer per month: This is the total number of visits by a specific customer in a month. 

- Trial conversion rate: calculated as the number of trial customers who becomes a paying customer divided by the number of total trial members. 

These metrics are non-financial metrics that has a causal effect on financial metrics like Return on Investments (ROI), Return on Equity (ROE), or Earnings per Share (EPS). For instance, if NPS is high then it indicates that customers like using OC&Gym service, which would most likely lead them to returning again, benefiting sales, net income, and earnings per share.

The only way to improve pricing strategy in the long run is to simply assign a price, get feedback from customers and look at how it impacts churn rate, adjust accordingly, and repeat.
 
OC&Gym should slowly raise price in small increments and observe how a combination of metrics (churn rate, net income) changes in response to price change. Simply repeat the process until an inflection price point is discovered, a price point where churn rate increases significantly compared to historical change (and profitability drops).
 
Additionally, they should consider increasing pricing tiers (more than just "standard" and "flexible"). This increase the number of options, enabling the business to response to various customer affluence segments who have varying purchasing power and price sensitivities. The business can also experiment with different ideas such as a program where the utlities at the fitness center are limited in exchange for lower price. For instance, members may not use or access certain machines or areas, but gets a reduced payment rate. This may address the need for mass customers while not ruining the experience for premium customers by keeping them at the different area of the fitness center.

### 5.1.1 Data preprocessing
Data preprocessing is crucial to make sure our data contains sufficient data and attributes for data analysis. Here, we create a new customer dataframe base on customer vistation detail and subscription profile for later analysis use.

```python
import pandas as pd
import matplotlib.pyplot as plt

import numpy as np
import seaborn as sns

import datetime as dt
from datetime import date
from dateutil.relativedelta import relativedelta

import warnings
warnings.filterwarnings('ignore')
```

```python
# import and merge customer visitation data
visitation = pd.read_csv('visitationdataPART1.csv',parse_dates=[['visitdate', 'TimeofVisit']], infer_datetime_format= True)
tmp = pd.read_csv('visitationdataPART2.csv',parse_dates=[['visitdate', 'TimeofVisit']], infer_datetime_format= True)
visitation = visitation.append(tmp,ignore_index=True)
visitation.index = visitation.CustomerID.values
visitation.head(3)
```

```python
# import customer subscription record
date_parser = lambda c: pd.to_datetime(c, format='%b-%y', errors='coerce')
subscription = pd.read_csv('subscriptionsdataFINAL.csv', parse_dates=['Join Date','End Date'], date_parser=date_parser)
subscription.index = subscription.CustomerID.values
subscription.head(3)
```

The dataset as of 2018-12 has customer records from 2009 to 2018.

```python
# new column count total peak and non-peak visit
tmp = pd.pivot_table(visitation, index='CustomerID',columns='peakvisits',aggfunc='count')
customer = subscription.join(tmp)
customer = customer.rename(columns={('visitdate_TimeofVisit', False): "count_non_peakvisit_total", ('visitdate_TimeofVisit', True): "count_peakvisit_total"})
```

```python
# new column 'count visit total' where equals to 'count_non_peakvisit_total' + 'count_peakvisit_total'
customer['count visit total'] = customer['count_non_peakvisit_total'] + customer['count_peakvisit_total'] 
```

```python
# set affected by price increase = True for customers left at 2015-12-01, just before price increase
customer['affected by price increase']=False
customer['affected by price increase'][customer['End Date'].dt.date == dt.date(2015, 12, 1)] = True
```

```python
# membership period = end date - join date
customer['membership period'] = customer['End Date'] - customer['Join Date']

# For Join Date = null : use 2018-12-01 to count the period
tmp = customer.copy()
tmp['today'] = pd.to_datetime(dt.date(2018, 12, 1))
tmp = tmp[tmp['membership period'].isna()]
tmp['membership period'] = tmp['today'] - tmp['Join Date']
customer['membership period'][customer['membership period'].isna()] = tmp['membership period']
customer['membership period'] = pd.to_timedelta(customer['membership period'], errors='coerce').dt.days
```

#### Add columns for recency of visits
A customer who visits a gym less per month or a once frequent user who has not come to the gym recently could be a candidate of likely-churn member. The business can acquire this information to identify ways to get members to use their membership more frequently.

Therefore, we add few more columns to record members recency of visiting for churn prediction. If end date is null, the customer is still active, we use 2018-12-01 as 'end date' for these records. 

We need the visit count 1 month before end date (the same as last month visit if the customer is still active), end date -3 and -6 month so that we can see the trend of customer visitation. From here, we can compute the delta percentage of 1 and 6 month, as well as 1 and 3 month before end date. If the delta percentage become more negative, it could be a sign of customer churn and the business may need to approach and keep them.

```python
# merging end date to visitation data
# If end date is null, the customer is still active, we use 2018-12-01 as 'end date' for these records.

visitation = visitation.join(customer['End Date'])
visitation['End Date'][visitation['End Date'].isna()] = pd.Timestamp('2018-12-01')
visitation = visitation.join(customer['Join Date'])
```

```python
# Add First month visit count to customer dataframe
tmp = visitation[visitation['visitdate_TimeofVisit'].dt.to_period('M') == visitation['Join Date'].dt.to_period('M') ]
tmp = pd.pivot_table(tmp, index='CustomerID',aggfunc='count')
tmp = tmp.drop(columns=['peakvisits','visitdate_TimeofVisit','End Date'])
tmp = tmp.rename(columns={"Join Date": "First month visit count"})
customer = customer.join(tmp)
```

### 5.1.4 Individual Churn prediction - Supervised Machine learning 
From the data analysis and member segment analysis, we found that 'First month visit count','standard subs' and 'membership period' together could be a good predictor for churn prediction modelling for individual customer. Regardless their gender, higher first month visit of standard member generally stay active for a longer time period. 

Now we can finally build an Individual Churn prediction model using that three attributes, and then predict whether the currently active member will churn.

Below, we train the kNN model using train dataset, then find the best value of k using the validation set, followed by predicting the individual churn using the test dataset. Although all k values give a very high accuracy of above 97%, we prefer lower k value that won't predict everyone churn. Afterall, we want to know whether customers who are currently active will churn, we predict this using the trained model. 

#### Model training

```python
ml = customer_not_affected_by_price[['CustomerID','First month visit count','standard subs','membership period','churn']]
ml.head(3)

# index = customer ID for later mapping
```

```python
import numpy as np
import pandas as pd
from scipy import stats

from sklearn import metrics #Import scikit-learn metrics module for accuracy calculation
from sklearn.model_selection import train_test_split # Import train_test_split function
```

```python
# kNN Classifier
from sklearn.neighbors import KNeighborsClassifier
```

```python
import matplotlib.pyplot as plt  
from sklearn.metrics import plot_confusion_matrix
from sklearn.metrics import classification_report

#Import scikit-learn metrics module for accuracy calculation
from sklearn import metrics
```

```python
# split data into training (60%) , validation (20%) and test (20%) sets

# shuffle and split into a training set (60%) and the rest (40%).
df_train, tmp = train_test_split(ml.iloc[:,1:], test_size = 0.4, shuffle = True)

# shuffle and split into a validation set (50%) and a test set (50%).
df_val, df_test = train_test_split(tmp, test_size = 0.5, shuffle = True)

# split data to inputs and output
X_train, y_train = df_train.iloc[:,:-1].replace(np.nan, 0), df_train.iloc[:,-1].replace(np.nan, 0)
X_val, y_val = df_val.iloc[:,:-1].replace(np.nan, 0), df_val.iloc[:,-1].replace(np.nan, 0)
X_test, y_test = df_val.iloc[:,:-1].replace(np.nan, 0), df_val.iloc[:,-1].replace(np.nan, 0)

X_train.head(3)
```

```python
# Z-score normalisation
X_train_zscore = (X_train - X_train.mean())/X_train.std(ddof=0)
X_val_zscore = (X_val - X_train.mean())/X_train.std(ddof=0)
X_test_zscore = (X_test - X_train.mean())/X_train.std(ddof=0)

X_train_zscore.head(3)
```

#### Choose of k 
All k gives very high accuracy rate, the differences are the number of false positives and false alarms. We always prefer false alarms but it doesn't make sense if everyone are predicted to churn, so a very big value of k is not preferred. k=5 would be the best for the churn prediction because it correctly predicts all churn and keep acceptable rate of false alarm.

```python
# trains the k-Nearest Neighbours classifiers for k = 1, 3, 5, 7, 10, 30
# Create KNN Classifier
k = [1, 3, 5, 7, 10, 30]
acc=[]
for i in k:
    knn = KNeighborsClassifier(n_neighbors=i)
    
    #Train the model using the training sets
    knn.fit(X_train_zscore, y_train)

    #Predict the response for validate dataset
    y_val_pred = knn.predict(X_val_zscore)
    
    # Model Accuracy, how often is the classifier correct?    
    print('k=',i)
    accuracy = metrics.accuracy_score(y_val, y_val_pred)
    print("Accuracy:",accuracy)
    acc.append(accuracy)
    
    #plot confusion matrix
    plot_confusion_matrix(knn, X_val_zscore, y_val)  
    plt.show()  
    
```

#### High accuracy on test dataset

```python
# predicts the generalisation error using the test data set
best_k = 5

# predict test output
knn = KNeighborsClassifier(n_neighbors=best_k)
knn.fit(X_train_zscore, y_train)
y_test_pred = knn.predict(X_test_zscore)

#plot confusion matrix
print("Testing dataset accuracy:",metrics.accuracy_score(y_test, y_test_pred))
print('Others：\n', classification_report(y_test, y_test_pred))
plot_confusion_matrix(knn, X_test_zscore, y_test)  
plt.show()
```

#### How likely will currently active member leave?

```python
X_test_active_member = ml[ml['churn'] == 0].iloc[:,1:-1].replace(np.nan, 0)
y_test_active_member = ml[ml['churn'] == 0].iloc[:,-1].replace(np.nan, 0)
X_test_active_member_zcore = (X_test_active_member - X_train.mean())/X_train.std(ddof=0)

# predict test output
knn = KNeighborsClassifier(n_neighbors=best_k)
knn.fit(X_train_zscore, y_train)
y_test_active_member_pred = knn.predict(X_test_active_member_zcore)

# percentage of customer churn
print('percentage of customer churn:', y_test_active_member_pred.sum()/(len(y_test_active_member_pred)))
```

The model predicts that almost 90% of customers are going to leave. The business should probably take active action to these customers such as giving promotions or introducing other plans to them to retain them.

```python
# group members by TimeofVisit
def visit_time(tv):
    if tv >= datetime.time(6,0) and tv <= datetime.time(11,59):
        return 'morning'
    elif tv >= datetime.time(12,0) and tv <= datetime.time(17,59):
        return 'afternoon'
    elif tv >= datetime.time(18,0) and tv <= datetime.time(20,59):
        return 'evening'
    elif tv>= datetime.time(21,0) and tv <= datetime.time(23,59):
        return 'night'
    elif tv>= datetime.time(0,0) and tv <= datetime.time(5,59):
        return 'latenight'
```

```python
# merge visitationdataPART1 and visitationdataPART1 into vad_merge
vad_merge = pd.concat([vad_1, vad_2])

# set index of vad_merge as CustomerID
vad_merge = vad_merge.set_index('CustomerID')

# change format of TimeofVisit into Year-Month-Day format
vad_merge['TimeofVisit'] = [datetime.time(int(dt.split(':')[0]),int(dt.split(':')[1])) for dt in vad_merge['TimeofVisit']]

# group TimeofVisit
vad_merge['GroupofVisitTime'] = [visit_time(tv) for tv in vad_merge['TimeofVisit']]    

# get dummy variables for GroupofVisitTime
vad_merge = pd.get_dummies(vad_merge, columns=['GroupofVisitTime'])

# replace True by 1 and False by 0 
vad_merge['is_peak'] = [1 if x else 0 for x in vad_merge['peakvisits']]

# generate visitactiondata 
# vad_merge.to_csv(r'C:\Users\surface pro4\Downloads\AIB\A1\visitactiondata.csv')
```
